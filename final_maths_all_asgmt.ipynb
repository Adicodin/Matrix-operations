```python
def det(m): # only for 2x2 and 3x3
    if len(m) == 4:
      det = m[0]*m[3] - m[1]*m[2]
    else:
      det = (m[0] * (m[4] * m[8] - m[5] * m[7])) - (m[1] * (m[3] * m[8] - m[5] * m[6])) + (
                  m[2] * (m[3] * m[7] - m[4] * m[6]))
      # print("determinant = ", det)
    return det

def minor(m):  # for 3x3 only it calculates -> minor sum and not minor
  try:
    sum_min = ((m[4] * m[8] - m[5] * m[7])) + ((m[0] * m[8] - m[2] * m[6])) + ((m[0] * m[4] - m[1] * m[3]))
    # print("sum of minors of diagonals = ", sum_min)
    return sum_min
  except:
    print("Sorry the minor sum can only be calculated for 3x3 matrix")
  


def ratio(lst):  # to print ratio in : form inputting a list of numbers
    ratio = ''
    for i in lst:
        if not i:
            ratio += (str(1)+':')
        else:
            num = 1
            for j in i:
                num *= j
            ratio += (str(num)+':')
    return (ratio[:-1])

#function for cubic eqn calculation
def cubic_root(a, b, c, d): # ax^3 + bx^2 + cx + d  here a, b, c, d should be = 1, -S1, S2, -S3
    print("The cubic eqn = λ^3 -{}λ^2 + {}λ -{}".format(abs(b), abs(c), abs(d)))
    # to find first root
    r1 = 0
    cubic_eqn = lambda x : a*x**3 + b*x**2 + c*x + d
    for i in [-2, -1, 1, 2]:
        if cubic_eqn(i) == 0:
            r1 = i
            break
    # print(r1)
    # to find other roots by r1
    # let the other quad equation formed be px^2 + qx + r
    p = 1
    q = b+r1
    r = c + q*r1
    D = q**2 - 4*p*r
    if D >= 0 and r1 != 0:
        r2 = (-q + (q ** 2 - 4 * p * r) ** (1 / 2)) / (2 * p)
        r3 = (-q - (q ** 2 - 4 * p * r) ** (1 / 2)) / (2 * p)
        # print("The roots of given cubic eqn are = {}, {}, {}".format(float(r1), r2, r3))
        int_roots = []
        for i in r1, r2, r3:
            if int(i) == i:
                int_roots.append(int(i))
        return int_roots
    else:
        print("The roots are not real")

# function to calculate HCF
def HCF(*lst):
  min_num = min(lst)
  if min_num == 0:
    return 0
  for i in range(1, min_num+1):
    cnt = 0
    for j in lst:
      if j%i == 0:
        cnt += 1
    if cnt == len(lst):
      hcf = i
  # print(hcf)
  return hcf


#function for eigen values
def egvl(m):
  if len(m) == 4:  # for 2x2 matrix
    a = 1
    b = -(m[0]+m[3])  # sum of diagonals S1
    c = m[0]*m[3]-m[1]*m[2]  # determinant S2 of 2x2
    #find roots of equation x and y
    x = (-b+(b**2-4*a*c)**(1/2))/(2*a)
    y = (-b-(b**2-4*a*c)**(1/2))/(2*a)
    eqn = 'λ^2 - {}λ + {}'.format(abs(b), c)
    print("The equation =", eqn)
    print("Eigen values are = ", int(x), ',', int(y))
    return int(x), int(y)
  else:  # for 3x3 matrix
    a = 1
    b = -(m[0]+m[4]+m[8]) # sum of diagonals S1
    c = minor(m)  # sum of minor of diagonals S2
    d = -det(m)  # determinant
    roots = cubic_root(a, b, c, d)  # storing cubic eqn roots
    roots_for_printing = ''  # variable used to print the roots
    for i in roots:
      roots_for_printing += (str(i)+',')
    print("Roots of the above eqn =", roots_for_printing[:-1])
    return roots
  
def egvt(m):
  if len(m) == 4:  # for 2x2 matrix
    x1, x2 = egvl(m)
    for root in x1, x2: # for modifying matrix with each root
      coef_y = m[1]
      coef_x = (m[0]-root)
      new_mat = [m[0]-root, m[1], m[2], m[3]-root]
      print("New matrix = ", new_mat, "with root =", root)
      hcf = HCF(abs(coef_y), abs(coef_x))
      if hcf != 0:
        lst = [[-coef_y//hcf], [coef_x//hcf]]
      else:
        lst = [[-coef_y], [coef_x]]
      egvt = ratio(lst)
      # printing the ratio or eigen vector
      print("The eigen vectors (the ratio) = ", egvt, sep='')
  else:
    cubic_roots = egvl(m)
    # for removing duplicate roots
    croot = []  # croot = cubic root
    for i in cubic_roots:
      if i not in croot:
        croot.append(i)
    for root in croot:
      new_mat = [m[i]-root if i%4 == 0 else m[i] for i in range(len(m))]
      print("New matrix =", new_mat, "with root =", root) 
      # found 3 variables' coefficient using cramers rule below
      coef_x = new_mat[1]*new_mat[5] - new_mat[2]*new_mat[4]
      coef_y = -(new_mat[0]*new_mat[5]-new_mat[2]*new_mat[3])
      coef_z = new_mat[0]*new_mat[4] - new_mat[1]*new_mat[3]
      coef_lst = [coef_x, coef_y, coef_z]
      print("ratio by cramers rule = ", coef_lst)
      hcf = HCF(abs(coef_x), abs(coef_y), abs(coef_z))
      if hcf != 0:
        lst = [[coef_x//hcf], [coef_y//hcf], [coef_z//hcf]]
      else:
        lst = [[coef_x], [coef_y], [coef_z]]
      egvt = ratio(lst)
      print("eigen vectors (ratio) = ", egvt)


mat = input("Enter the numbers of matrices in one line with spaces = ")
mat = [int(i) for i in (mat).split(' ')]
ch = int(input("1. Find determinant\n2. Find sum of minors of diagonals\n3. Find the eigen values and eigen vectors\nEnter your choice = "))
if ch == 1:
    d = det(mat)
    print("determinant =", d)
elif ch == 2:
    m = minor(mat)
    if m:
      print(m)
elif ch == 3:
    egvt(mat)
```

```python
def lst_num(l1, l2):  # to multiply elements of nested list
    l = []
    for i in range(len(l1)):
        l.append(l1[i]*l2[i])
    return l


mat_inp = input("Enter the matrix A numbers in one line with spaces = ")
A = ([int(i) for i in mat_inp.split(' ')])
mat_inp = input("Enter the matrix B numbers in one line with spaces = ")
B = ([int(i) for i in mat_inp.split(' ')])
div = int(len(A)**0.5)
Arows, Bcols = [], []
for i in range(div):  # to store matrix as per rows and cols
    r = A[i*div : i*div + div]
    Arows.append(r)
for i in range(div):
    c = B[i::div]
    Bcols.append(c)
# print(Arows)
# print(Bcols)
AB = []
for i in Arows:
    for j in Bcols:
        mul = lst_num(i, j)
        AB.append(sum(mul))

print("Multiplied matrix =", AB)
```
